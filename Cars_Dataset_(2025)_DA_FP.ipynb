## Stage 2 – Data Cleaning and Pre-processing

Handle missing values (impute or drop)

Handle duplicates

Treat outliers if required

Check skewness and apply transformations

Convert data types if needed

Feature transformations (date parts, derived fields if required for analysis)

```python
# Impute missing values in 'CC/Battery Capacity', 'Performance(0 - 100 )KM/H', and 'Torque' with the mode grouped by 'Company Names'
for col in ['CC/Battery Capacity', 'Performance(0 - 100 )KM/H', 'Torque']:
    df[col] = df.groupby('Company Names')[col].transform(lambda x: x.fillna(x.mode()[0] if not x.mode().empty else x.fillna(x.mode())))

print("✅ Imputation completed using Mode + Group-wise Mode.")
```

```python
# Display duplicate records for review
df[df.duplicated()].head()
```

```python
# Drop duplicate rows
df.drop_duplicates(inplace=True)

print("✅ Duplicates removed successfully.")
print("Remaining records:", df.shape[0])
```

```python
# Detect and remove outliers using IQR for numeric columns
numeric_cols = df.select_dtypes(include=['int64', 'float64']).columns

for col in numeric_cols:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    outliers = df[(df[col] < lower_bound) | (df[col] > upper_bound)]
    print(f"{col}: {len(outliers)} outliers")

print("✅ Outlier detection completed.")
```

```python
# Clean and convert 'Cars Prices' to numeric
def clean_price(price):
    price = str(price).replace('$', '').replace(',', '').strip()
    if ' - ' in price:
        try:
            low, high = price.split(' - ')
            return (float(low) + float(high)) / 2
        except ValueError:
            return np.nan
    elif ' ' in price: # Handle cases like '€ 33,000'
         try:
            parts = price.split(' ')
            return float(parts[-1])
         except ValueError:
            return np.nan
    else:
        try:
            return float(price)
        except ValueError:
            return np.nan

df['Cars Prices_cleaned'] = df['Cars Prices'].apply(clean_price)
```

```python
# Clean and convert 'Performance(0 - 100 )KM/H' to numeric
def clean_performance(performance):
    performance = str(performance).replace(' sec', '').strip()
    if ' - ' in performance:
        try:
            low, high = performance.split(' - ')
            return (float(low) + float(high)) / 2
        except ValueError:
            return np.nan
    else:
        try:
            return float(performance)
        except ValueError:
            return np.nan

df['Performance_cleaned'] = df['Performance(0 - 100 )KM/H'].apply(clean_performance)
```

# Univariate Analysis

---





```python
# Distribution of Car Prices
plt.figure(figsize=(10, 6))
sns.histplot(df['Cars Prices_cleaned'].dropna(), kde=True, bins=50)
plt.title('Distribution of Car Prices')
plt.xlabel('Price (in USD)')
plt.ylabel('Count')
plt.show()
```

### Budget vs. Premium
*   Most cars are priced in the mid-range, indicating a competitive
segment.
*   The tail towards higher prices shows a few premium/luxury models.
*   Helps identify market segments — budget vs. premium.








```python
# Car Count by Fuel Type
plt.figure(figsize=(6,4))
sns.countplot(data=df, x='Fuel Types', order=df['Fuel Types'].value_counts().index)
plt.title('Car Count by Fuel Type')
plt.xlabel('Fuel Type')
plt.ylabel('Number of Cars')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

### Car Count by Fuel Type

*   Majority of cars use Petrol or Diesel, while EV adoption is emerging.

*   Shows how fuel infrastructure and consumer preference affect demand



```python
plt.figure(figsize=(6,4))
sns.boxplot(x=df['CC/Battery Capacity'])
plt.title('Distribution of Engine/Battery Capacity')
plt.xlabel('CC / Battery Capacity')
plt.show()
```

### Boxplot of Engine Capacity (CC/Battery Capacity)


*  Most vehicles cluster around mid-level engine capacities.
*  Outliers represent high-performance or premium vehicles.





```python
plt.figure(figsize=(10,5))
top_brands = df['Company Names'].value_counts().head(10)
sns.barplot(x=top_brands.index, y=top_brands.values)
plt.title('Top 10 Car Brands by Model Count')
plt.xlabel('Company')
plt.ylabel('Number of Models')
plt.xticks(rotation=45)
plt.show()
```

### Top 10 Car Companies by Model Count

*   Shows market dominance of key manufacturers.

*  These brands likely have diverse product portfolios across price ranges.




# Bivariate Analysis

---



```python
#Average Price by Fuel Type
plt.figure(figsize=(10, 6))
sns.barplot(data=df, x='Fuel Types', y='Cars Prices_cleaned', estimator='mean', errorbar=None)
plt.title('Average Price by Fuel Type')
plt.xlabel('Fuel Type')
plt.ylabel('Average Price (USD)')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
#Price vs Engine Capacity
plt.figure(figsize=(12,4))
sns.scatterplot(data=df, x='CC/Battery Capacity', y='Cars Prices_cleaned', hue='Fuel Types')
plt.title('Price vs Engine/Battery Capacity')
plt.xlabel('Engine/Battery Capacity')
plt.ylabel('Price')
plt.show()
```

### Price vs Engine Capacity


*   Positive relationship — higher engine capacity → higher price.

*   EVs form separate clusters, showing distinct pricing strategy.



```python
# Price vs Performance
plt.figure(figsize=(8,9))
sns.scatterplot(data=df, x='Performance_cleaned', y='Cars Prices_cleaned', hue='Company Names')
plt.title('Price vs Performance (0-100 KM/H)')
plt.xlabel('Performance Time (sec)')
plt.ylabel('Price')
plt.show()
```

### Price vs Performance


*   Faster acceleration (lower 0–100 time) often correlates with higher prices.

*   Indicates performance premium among sports or luxury cars.



# Multivariate Analysis

---





```python
#Company × Fuel Type vs Average Price
plt.figure(figsize=(12,6))
grouped = df.groupby(['Company Names','Fuel Types'])['Cars Prices_cleaned'].mean().reset_index()
sns.barplot(data=grouped, x='Company Names', y='Cars Prices_cleaned', hue='Fuel Types')
plt.title('Average Price by Company and Fuel Type')
plt.xlabel('Company')
plt.ylabel('Average Price')
plt.xticks(rotation=45)
plt.legend(title='Fuel Type')
plt.show()
```



*   Premium brands (e.g., BMW, Mercedes) show high EV/Hybrid prices.

*   Mid-range brands dominate petrol/diesel segments.

*   Demonstrates brand–fuel strategy positioning.
